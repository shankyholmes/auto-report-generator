```python
import pandas as pd
```

```python
import scipy.interpolate as sci
```

```python
import numpy as np
```

```python
import sys
```

```python
import csv
```

```python
df = pd.read_excel("MPM_NO.xlsx")
```

```python
df
```

```python
df.drop(0,inplace=True)
```

```python
df2 = df.dropna(subset = ['AVL_INDEP_TIME.1'])
```

```python
df3 = df2.drop(df2[df2['SPEED'] == 'SPEED'].index)
```

```python
df3
```

```python
df3['G_SOOT'].iloc[100:155]
```

```python
cols = df3.select_dtypes(exclude=['float']).columns
df3[cols] = df3[cols].apply(pd.to_numeric, downcast='float', errors='coerce')
```

```python
df3["G_SOOT"] = df3["G_SOOT"].fillna(0)
```

```python
df3['soot_mg_s'] = df3['G_SOOT']*(1000/3600)
```

```python
df3['soot_mg_s'].iloc[100:155]
```

```python
x = df3['SPEED'].to_numpy()
y = df3['M_FUEL'].to_numpy()
z= df3['soot_mg_s'].to_numpy()
z2 = df3['G_SOOT'].to_numpy()
```

```python
sys.getsizeof(z2)
```

```python
f = sci.interp2d(x,y,z,kind = 'linear',bounds_error = False, fill_value = 0)
```

```python
x_new = np.arange(1000,2300,100)
y_new = np.arange(0,11,1)
```

```python
print(y_new)
```

```python
out = f(x_new,y_new)
```

```python
with open("out.csv","w+") as my_csv:
    csvWriter = csv.writer(my_csv,delimiter=',')
    csvWriter.writerows(out)
```
